```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0,2*np.pi)
r = 1 + np.sin(theta)

fig = plt.figure()
ax = fig.add_subplot(111, projection="polar")
ax.plot(theta,r)

plt.show()
```
